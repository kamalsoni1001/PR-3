# Data Analysis of Excel File

This notebook:
- Checks data structure
- Categorizes columns into numeric and categorical
- Checks missing values
- Creates a summary table

```python
import pandas as pd

# Load the Excel file
file_path = "student_performance.xlsx"
df = pd.read_excel(file_path)

# -------------------------------
# 1. Data Structure Overview
# -------------------------------
print("===== DATA STRUCTURE =====")
df.info()

print("\nFirst 5 Rows:")
df.head()
```

```python
# -------------------------------
# 2. Categorize Columns
# -------------------------------
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print("Numeric Columns:", numeric_cols)
print("Categorical Columns:", categorical_cols)
```

```python
# -------------------------------
# 3. Missing Values Check
# -------------------------------
missing_values = df.isnull().sum()
print("\n===== MISSING VALUES =====")
print(missing_values)
```

```python
# -------------------------------
# 4. Data Categorization Summary Table
# -------------------------------
summary_df = pd.DataFrame({
    'Column Name': df.columns,
    'Data Type': df.dtypes.values,
    'Category': ['Numeric' if col in numeric_cols else 'Categorical' for col in df.columns],
    'Missing Values': df.isnull().sum().values,
    'Unique Values': df.nunique().values
})

print("\n===== DATA SUMMARY TABLE =====")
summary_df
```
